```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

seeds_df = pd.read_csv('./seeds-less-rows.csv')

# remove the grain species from the DataFrame, save for later
varieties = list(seeds_df.pop('grain_variety'))

# extract the measurements as a NumPy array
samples = seeds_df.values
```

```python
seeds_df.head()
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
#calculate the mergings
mergings = linkage(samples, method='complete')

#to plot dendogram, we use varieties as labels
plt.figure(figsize=(15, 5))
dendrogram(mergings, labels=varieties, leaf_rotation=90, leaf_font_size=6),

plt.show()
```

```python
fn = './company-stock-movements-2010-2015-incl.csv'
stocks_df = pd.read_csv(fn, index_col=0)

companies = list(stocks_df.index)
movements = stocks_df.values
```

```python
stocks_df.head()
```

```python
from sklearn.preprocessing import normalize
# normalizing the movements
normalized_movements = normalize(movements)

#to calaculate linkage we have to perform mergings
mergings = linkage(normalized_movements, method='complete')

#to plot dendrogram
plt.figure(figsize=(15, 5))
dendrogram(mergings, labels=companies, leaf_rotation=90, leaf_font_size=10)

plt.show()
```

```python
seeds_df = pd.read_csv('./seeds-less-rows.csv')

# remove the grain species from the DataFrame, save for later
varieties = list(seeds_df.pop('grain_variety'))

# extract the measurements as a NumPy array
samples = seeds_df.values

from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

mergings = linkage(samples, method='complete')

dendrogram(mergings,
           labels=varieties,
           leaf_rotation=90,
           leaf_font_size=6,
)
plt.show()
```

```python
from scipy.cluster.hierarchy import fcluster

# using fcluster to extract labels
labels = fcluster(mergings, 6, criterion='distance')

#create a dataframe
df = pd.DataFrame({'labels': labels, 'varieties':varieties})

#creat a crosstab
ct = pd.crosstab(df['labels'], df['varieties'])

print(ct)
```
